```python
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline

matches_count = 100

# read images
img1 = cv.imread('pickaxe_with_bg.png')
img2 = cv.imread('inventory_fg.png')


img1 = cv.cvtColor(img1, cv.COLOR_BGR2RGB)
img2 = cv.cvtColor(img2, cv.COLOR_BGR2RGB)
```

```python
#sift
sift = cv.xfeatures2d.SIFT_create()

keypoints1, descriptors1 = sift.detectAndCompute(img1, None)
keypoints2, descriptors2 = sift.detectAndCompute(img2, None)

#feature matching
bf = cv.BFMatcher.create()
matches = bf.match(descriptors1, descriptors2)
matches = sorted(matches, key = lambda x: x.distance)

plt.imshow(cv.drawKeypoints(img1, keypoints1, None, flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS))
plt.savefig("plotsOpenCV/siftKeypoints.png", bbox_inches="tight")
plt.show()
plt.imshow(cv.drawMatches(img1, keypoints1, img2, keypoints2, matches[:matches_count], None, flags=2))
plt.savefig("plotsOpenCV/siftMatches.png", bbox_inches="tight")
plt.show()
```

```python
brisk = cv.BRISK_create()

# Find the keypoints and compute the descriptors for input and training-set image
keypoints1, descriptors1 = brisk.detectAndCompute(img1, None)
keypoints2, descriptors2 = brisk.detectAndCompute(img2, None)

#feature matching
bf = cv.BFMatcher.create()
matches = bf.match(descriptors1, descriptors2)
matches = sorted(matches, key = lambda x: x.distance)

plt.imshow(cv.drawKeypoints(img1, keypoints1, None, flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS))
plt.savefig("plotsOpenCV/BriskKeypoints.png", bbox_inches="tight")
plt.show()
plt.imshow(cv.drawMatches(img1, keypoints1, img2, keypoints2, matches[:matches_count], None, flags=2))
plt.savefig("plotsOpenCV/briskMatches.png", bbox_inches="tight")
plt.show()
```

```python
# Initiate Freak descriptor
freak = cv.xfeatures2d.FREAK_create()

# Find the keypoints and compute the descriptors for input and training-set image
keypoints1, descriptors1 = freak.compute(img1, keypoints1)
keypoints2, descriptors2 = freak.compute(img2, keypoints2)

#feature matching
bf = cv.BFMatcher.create()
matches = bf.match(descriptors1, descriptors2)
matches = sorted(matches, key = lambda x: x.distance)

plt.imshow(cv.drawKeypoints(img1, keypoints1, None, flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS))
plt.savefig("plotsOpenCV/freakKeypoints.png", bbox_inches="tight")
plt.show()
plt.imshow(cv.drawMatches(img1, keypoints1, img2, keypoints2, matches[:matches_count], None, flags=2))
plt.savefig("plotsOpenCV/freakMatches.png", bbox_inches="tight")
plt.show()
```
